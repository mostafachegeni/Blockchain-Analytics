```python
# Create "Merged Parents Array" and define required methods:


print('----------------------')

# ct stores current time
ct = datetime.datetime.now()
print("current time: ", ct)


##########################################################################################
parents_merged = np.array([0] * unique_addresses_len)

for i in range(len(parents_merged)):
    parents_merged[i] = i


clustering_array = np.array([0] * unique_addresses_len)

##########################################################################################
def parent (id1, parents_array):
    return parents_array[id1];

##########################################################################################
def find_parent (id1, parents_array):
    while (id1 != parent(id1, parents_array)):
        new_parent = parent(parent(id1, parents_array), parents_array);
        id1        = new_parent;
    
    return id1

##########################################################################################
# Define link_address method:
def link_address (addr_position_1, addr_position_2, parents_array):
    id1 = find_parent(addr_position_1, parents_array)
    id2 = find_parent(addr_position_2, parents_array)

    if (id1 == id2):
        return

    # swap if id1 < id2
    if id1 < id2:
        x = id1
        id1 = id2
        id2 = x

    parents_array[id1] = id2
    return

##########################################################################################
def resolveAll (parents_array):
    for id1 in range(len(parents_array)):
        parents_array[id1] = find_parent(id1, parents_array);
    return

##########################################################################################
def remapClusterIds (parents_array):
    cluster_count = 0
    place_holder = 9999999999999
    new_cluster_ids = [place_holder] * len(parents_array)
    
    for i in range(len(clustering_array)):
        clustering_array[i] = parents_array[i]

    for i in range(len(clustering_array)):
        parent_index = clustering_array[i]

        if (new_cluster_ids [parent_index] == place_holder):
            new_cluster_ids [parent_index] = cluster_count
            cluster_count = cluster_count + 1

        clustering_array[i] = new_cluster_ids [parent_index]

    return cluster_count;

##########################################################################################
def merge_parents(parent_array):
    length_ = len(parent_array)
    
    if (length_ != len(parents_merged)):
        print('merge_parents Error: -1 (Length)')
        return -1
    
    for i in range(len(parents_merged)):
        link_address (i, parent_array[i], parents_merged)


##########################################################################################
def BinarySearch(a, x):
    i = bisect_left(a, x)
    if i < len(a) and a[i] == x:
        return i
    else:
        print('BinarySearch Error: -1')
        return -1

##########################################################################################
def BinarySearch_Find_start_end(a, x):
    i = bisect_left(a, x)
    j = bisect_right(a, x) - 1
    if i < len(a) and a[i] == x and j < len(a) and a[j] == x:
        return [i, j]
    else:
        print('BinarySearch Error: -1')
        print('i = ', i)
        print('j = ', j)
        return -1



##########################################################################################



##########################################################################################
print('----------------------')
et = datetime.datetime.now() - ct
print("Total elapsed time (Fill Clustering Array with Heuristics): ", et)

##########################################################################################
print('----------------------')
print('done!')
```

```python
# Find number of members in each cluster:

print('----------------------')
ct = datetime.datetime.now()
print("start time: ", ct)




print('----------------------')
sorted_clustering_array = np.sort(clustering_array)
print('clustering_array was sorted !')


num_of_clusters = max(clustering_array) + 1
print('number of clusters = ', num_of_clusters)


num_of_cluster_members = np.array([0] * num_of_clusters)


for i in range(num_of_clusters):
    x = BinarySearch_Find_start_end(sorted_clustering_array, i)
    num_of_cluster_members[i] = x[1] - x[0] + 1
    #print('num_of_cluster_members[' + str(i) + '] = ', num_of_cluster_members[i]);




print('----------------------')
et = datetime.datetime.now() - ct
print("elapsed time: ", et)


##########################################################################################

print('----------------------')
print('done!')
```
